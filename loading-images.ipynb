## Images and Python

```python
from PIL import Image
```

```python
# Open image
im = Image.open('images/hmmm/two-thumbs-up.png')
display(im)
```

```python
# Rotate and open image
im = Image.open('images/hmmm/two-thumbs-up.png')
display(im.rotate(90))
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as img

im = img.imread('images/hmmm/two-thumbs-up.png')
plt.imshow(im)
plt.show()
```
